```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
df=pd.read_csv("/content/moviesData.csv")
df
```

```python
df.info()
```

```python
ldf=df[df['original_language']!='en']
ldf
```

```python
sb.countplot(x='original_language',data=ldf)
plt.xticks(rotation="vertical")
plt.show()
```

```python
df['release_date']=pd.to_datetime(df['release_date'])
```

```python
df['year']=df['release_date'].dt.year
```

```python
ydf=df[(df['year']>=2000) & (df['year']<=2015)]
ydf
```

```python
tyr=ydf.groupby('year')['revenue'].sum()
tyr
```

```python
sb.lineplot(x=tyr.index,y=tyr.values,data=tyr,marker="o")
```

```python
sb.lineplot(x='year',y='revenue',data=ydf,marker="o")
```

```python
hdf=df.nlargest(5,'revenue')
hdf
```

```python
sb.barplot(x='original_title',y='revenue',data=hdf)
```
